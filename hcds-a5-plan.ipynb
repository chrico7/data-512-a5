## Motivation / Problem Statement

In March of 2013, the Environmental Protection Agency (EPA) designated the City of Bainbridge Island aquifer system as a sole source aquifer (https://semspub.epa.gov/work/10/100006730.pdf). According to the EPA, this designation identifies areas with an aquifer that provides more than 50% of the drinking water consumed in the area overlying the aquifer and, if contaminated, "would create a significant hazard to public health".

Since receiving this designation, the City of Bainbridge Island has experienced significant growth and development; and rumors about diminishing water supply, saltwater intrusion, and pollution are rampant.

As a Bainbridge Island resident and native, the health of the water supply is of critical importance to me. My intention with this project is to perform a statistical analysis of aquifer health by investigating well log data from the Washington State Department of Ecology. I will then send this analysis to members of the Bainbridge Island City Council to provide them with additional information as they attempt to craft policy.

## Data Used

To perform this analysis, I will be using Well Logs data from the Washington State Department of Ecology. This dataset, described at the WA DOE Wells homepage (https://ecology.wa.gov/Water-Shorelines/Water-supply/Wells) and downloadable by using the Washington State Well Report Viewer Text Search utility (https://fortress.wa.gov/ecy/wellconstruction/map/WCLSWebMap/textsearch.aspx) contains records for each recorded well in the state. Specifically, the dataset contains the well location, diameter, depth, and completion date.

Because this is public data there is no license required. However, the terms of use clearly state that the DOE makes no claim to the accuracy of the data, is not liable for any use or misuse of the data, and that state law prevents private citizen names and addresses from being used for solicitation or commercial purposes.

I chose to use this dataset because drilling a new well is expensive and my belief is that the well will be drilled only as deep as it needs to be to provide water. Therefore, the depth of these wells over time should provide us with a good estimate of the variation in aquifer levels.

The main ethical consideration of using this dataset is that it contains the names of private citizens and well location data that could be used to find someone's residence. Since my analysis does not need the owner name, I will scrub this field from my final data.

## Background

In March of 2013, the Environmental Protection Agency (EPA) designated the City of Bainbridge Island aquifer system as a sole source aquifer (https://semspub.epa.gov/work/10/100006730.pdf). As part of this designation process, the EPA produced the Support Document for Sole Source Aquifer Designation of the Bainbridge Island Aquifer System (https://semspub.epa.gov/work/10/100006730.pdf). This document contains significant background information on the geography, hydrogeology, aquifer characteristics, risks of contamination, and population characteristics of Bainbridge Island. For the purposes of this study, the most important content in the Support Document is the specific description of the six Bainbridge Island aquifers; including their approximate location, depth, and percentage of supported wells. 

- Perched Aquifer System: > 200' above MSL
- Semi-Perched Aquifer: 100' to -20' MSL
- Sea Level Aquifer: 40' to -230' MSL
- Glaciomarine Aquifer System: -76' to -400' MSL
- Fletcher Bay Aquifer System: -690' to -1010' MSL
- Bedrock Aquifer System

An additional resource is the Groundwater Management/Monitoring Program run by the City of Bainbridge Island (https://www.bainbridgewa.gov/174/Groundwater-Management-Program). This program collects monthly measurements from 84 island wells and looks at the water level trend over time. 

I decided to perform this study because the politics of aquifer health on Bainbridge Island has become toxic. Aquifer preservation has become a weapon used by anti-growth activists in their attempt to halt any and all development. Conversely, those interests that profit from construction and development write off any concerns about aquifer health so they can complete their next project. Despite the significant amount of prior work from the USGS, EPA, and City of Bainbridge Island, there is still not a community consensus on how our water resources should be managed. I hope that this study will provide some additional information to the community.


## Methodology

This study is primarily based on the Well Logs data from the Washington Department of Ecology. This data is available in two places: as a csv download from the Text Viewer utility described above, or as a GIS download from the Research Data section of the DOE website (https://ecology.wa.gov/Research-Data/Data-resources/Geographic-Information-Systems-GIS/Data#w). The Well Logs data from both of these sources will be downloaded, compared for similarity, and if necessary, combined to yield the most complete dataset possible.

The minimum fields required are the date the well was drilled, the well depth, and the location of the well. Since the aquifer depths are measured off of mean sea level (MSL) RQ2 will likely require the approximate MSL of the well head. This will allow the well depth field to be normalized to mean sea level as well which will hopefully assist with well-aquifer mapping. Initial inspection of the data also suggests that there are a significant number of missing fields. Not only will I need to document and quantify missing data, but I will also need to ensure that there are a large enough number of samples evenly distributed across the island and between each aquifer.

Once the data has been gathered and cleaned, I will attempt to determine which aquifer each well is drawing from based on the recorded well depth. Because I do not have any labeled well data that maps wells to aquifers, I will need to use the unsupervised k-means clustering algorithm with k=6 (for the six different aquifers). Next, I will aggregate the well data to annual resolution by taking the mean of the well depth and diameter for each year. This will hopefully mitigate any seasonal differences in well drilling. Finally, I will plot the annual mean well depth for each aquifer and then perform a regression of well depth on time. I will then test the hypothesis that the regression coefficient is 0 with a 95% confidence level. 

Because the results of this study is intended for members of the public at large, the final presentation of these findings must be clearly understandable by a non-technical audience. Because of this restriction I have decided to use time-series plotting of the mean annual well depth data as well as mapping where possible. First, I will map all of the well records on Bainbridge Island colorized by the assigned aquifer. This will allow readers to visualize the amount and distribution of data used in the study. Next, I will create a chart that shows the number of wells and their depth over time for each aquifer. This will allow reader to see the volume of wells drilled over time as well as the association between depth and tapped aquifer. Finally, I will present the mean well depth trend for each aquifer and note any that showed a statistically significant trend.
